```python
import asyncio
from playwright.async_api import async_playwright
from bs4 import BeautifulSoup
import pandas as pd

async def run():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        # Intercept requests to debug API calls
        blocked_resources = [
      "ads.pubmatic.com", "sync.inmobi.com", "sync.adkernel.com",
      "tracker-shr.ortb.net", "sync.srv.stackadapt.com", "creativecdn.com",
      "pixel-eu.rubiconproject.com", "cms.quantserve.com"
  ]

        async def log_and_block(route, request):
            print("Request:", request.url)
            if any(domain in request.url for domain in blocked_resources):
                await route.abort()
            else:
                await route.continue_()

        await page.route("**/*", log_and_block)


        await page.goto("https://letterboxd.com/film/barbie/reviews/", wait_until="domcontentloaded", timeout=60000)

        # Wait for reviews to load
        try:
            await page.wait_for_selector("li.list-reviews__item", timeout=10000)
        except:
            print("Reviews did not load within 10 seconds.")

        # Scroll down to load more
        for _ in range(5):
            await page.mouse.wheel(0, 1000)
            await page.wait_for_timeout(2000)

        content = await page.content()

        with open("page.html", "w", encoding="utf-8") as f:
            f.write(content)

        await browser.close()

        soup = BeautifulSoup(content, "html.parser")
        reviews = []

        for block in soup.select("li.list-reviews__item"):
            text_elem = block.select_one("div.body-text")
            rating_elem = block.select_one("span.rating")
            review_text = text_elem.text.strip() if text_elem else None
            rating_class = None
            if rating_elem and "class" in rating_elem.attrs:
                for cls in rating_elem["class"]:
                    if cls.startswith("rated-"):
                        rating_class = cls
                        break
            reviews.append({
                "review": review_text,
                "rating": rating_class
            })

        print(f"Number of reviews scraped: {len(reviews)}")
        print("Sample review entry:", reviews[0] if reviews else "No reviews found")

        df = pd.DataFrame(reviews)

        def rating_to_stars(rating_class):
            if rating_class and rating_class.startswith("rated-"):
                return int(rating_class.split("-")[1]) / 10
            return None

        df["stars"] = df["rating"].apply(rating_to_stars)
        print(df.head())

await run()
```

```python
import asyncio
from playwright.async_api import async_playwright
from bs4 import BeautifulSoup
import pandas as pd

async def scrape_letterboxd_reviews(max_pages=5):
    base_url = "https://letterboxd.com"
    current_url = f"{base_url}/film/barbie/reviews/"
    reviews = []

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        for page_num in range(1, max_pages + 1):
            print(f"📄 Scraping page {page_num}: {current_url}")
            await page.goto(current_url, timeout=60000, wait_until="domcontentloaded")
            await page.wait_for_timeout(1000)  # let it settle

            content = await page.content()
            soup = BeautifulSoup(content, "html.parser")

            for review_body in soup.select("div.body"):
                rating_span = review_body.select_one("div.content-reactions-strip span.rating")
                rating = None
                if rating_span:
                    for cls in rating_span.get("class", []):
                        if cls.startswith("rated-"):
                            try:
                                rating = int(cls.split("rated-")[1]) / 2
                                break
                            except:
                                rating = None

                review_text_div = review_body.select_one("div.js-review div.body-text")
                review_text = review_text_div.text.strip() if review_text_div else None

                if review_text:
                    reviews.append({
                        "review": review_text,
                        "stars": rating
                    })

            # Try to find the next page
            next_page = soup.select_one("div.pagination a.next")
            if next_page and next_page.get("href"):
                current_url = base_url + next_page["href"]
            else:
                print("🚫 No more pages found.")
                break

        await browser.close()

    df = pd.DataFrame(reviews)
    print(df.head(10))
    print(f"✅ Total reviews scraped: {len(df)}")
    return df

# To run:
await scrape_letterboxd_reviews()
```

```python
import asyncio
from playwright.async_api import async_playwright
from bs4 import BeautifulSoup
import pandas as pd

# 1️⃣ Get slugs of popular films
async def get_popular_movie_slugs(n=10):
    slugs = []
    url = "https://letterboxd.com/films/popular/"
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto(url, timeout=60000, wait_until="domcontentloaded")
        await page.wait_for_timeout(1000)

        html = await page.content()
        soup = BeautifulSoup(html, "html.parser")

        for a in soup.select("ul.poster-list li div a[href^='/film/']"):
            href = a.get("href", "")
            slug = href.split("/")[2] if len(href.split("/")) > 2 else None
            if slug and slug not in slugs:
                slugs.append(slug)
            if len(slugs) >= n:
                break

        await browser.close()
    print(f"🎯 Found {len(slugs)} popular movie slugs.")
    return slugs


# 2️⃣ Scrape reviews for one movie
async def scrape_reviews_for_movie(slug, max_pages=3):
    base_url = "https://letterboxd.com"
    current_url = f"{base_url}/film/{slug}/reviews/"
    reviews = []

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        for page_num in range(1, max_pages + 1):
            print(f"📄 Scraping '{slug}' - Page {page_num}: {current_url}")
            await page.goto(current_url, timeout=60000, wait_until="domcontentloaded")
            await page.wait_for_timeout(1000)

            content = await page.content()
            soup = BeautifulSoup(content, "html.parser")

            for review_body in soup.select("div.body"):
                rating_span = review_body.select_one("div.content-reactions-strip span.rating")
                rating = None
                if rating_span:
                    for cls in rating_span.get("class", []):
                        if cls.startswith("rated-"):
                            try:
                                rating = int(cls.split("rated-")[1]) / 2
                                break
                            except:
                                rating = None

                review_text_div = review_body.select_one("div.js-review div.body-text")
                review_text = review_text_div.text.strip() if review_text_div else None

                if review_text:
                    reviews.append({
                        "movie": slug,
                        "review": review_text,
                        "stars": rating
                    })

            next_page = soup.select_one("div.pagination a.next")
            if next_page and next_page.get("href"):
                current_url = base_url + next_page["href"]
            else:
                print(f"🚫 No more pages found for '{slug}'.")
                break

        await browser.close()

    df = pd.DataFrame(reviews)
    print(df.head(3))
    print(f"✅ Total reviews scraped for '{slug}': {len(df)}")
    return df

# 3️⃣ Run everything
async def scrape_top_movies(n_movies=5, max_pages=3):
    all_reviews = pd.DataFrame()
    slugs = await get_popular_movie_slugs(n=n_movies)

    for slug in slugs:
        df = await scrape_reviews_for_movie(slug, max_pages=max_pages)
        df.to_csv(f"{slug}_reviews.csv", index=False)
        all_reviews = pd.concat([all_reviews, df], ignore_index=True)

    all_reviews.to_csv("popular_movies_reviews.csv", index=False)
    print(f"📦 Combined reviews saved to 'popular_movies_reviews.csv'.")

# ⏩ Start scraping
await scrape_top_movies(n_movies=5, max_pages=3)
```

```python
import asyncio
from playwright.async_api import async_playwright
from bs4 import BeautifulSoup
import pandas as pd
import random
from textblob import TextBlob

MOVIE_LIMIT = 20
REVIEWS_PER_MOVIE = 30
MAX_REVIEW_PAGES = 5


def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0.2:
        return "positive"
    elif polarity < -0.2:
        return "negative"
    else:
        return "neutral"


async def get_popular_movie_slugs(n=MOVIE_LIMIT):
    slugs = []
    url = "https://letterboxd.com/films/popular/"
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto(url, timeout=60000, wait_until="domcontentloaded")
        await page.wait_for_timeout(1000)

        html = await page.content()
        soup = BeautifulSoup(html, "html.parser")

        for a in soup.select("ul.poster-list li div a[href^='/film/']"):
            href = a.get("href", "")
            slug = href.split("/")[2] if len(href.split("/")) > 2 else None
            if slug and slug not in slugs:
                slugs.append(slug)
            if len(slugs) >= n:
                break

        await browser.close()
    print(f"🎯 Found {len(slugs)} movie slugs.")
    return slugs


async def scrape_movie_data(slug):
    base_url = "https://letterboxd.com"
    movie_url = f"{base_url}/film/{slug}/"
    reviews_url = f"{movie_url}reviews/"

    movie_data = {}
    reviews = []

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        # Scrape movie title and year
        try:
            await page.goto(movie_url, timeout=60000, wait_until="domcontentloaded")
            soup = BeautifulSoup(await page.content(), "html.parser")
            title_tag = soup.select_one("h1.headline-1")
            year_tag = soup.select_one("small.number")

            movie_data = {
                "slug": slug,
                "title": title_tag.text.strip() if title_tag else slug,
                "year": year_tag.text.strip("()") if year_tag else "",
            }
        except Exception as e:
            print(f"❌ Error scraping movie info for {slug}: {e}")
            return None, []

        # Scrape reviews (up to MAX_REVIEW_PAGES)
        current_url = reviews_url
        total_reviews_collected = 0
        for page_num in range(1, MAX_REVIEW_PAGES + 1):
            await page.goto(current_url, timeout=60000, wait_until="domcontentloaded")
            await page.wait_for_timeout(1000)
            soup = BeautifulSoup(await page.content(), "html.parser")

            for review_body in soup.select("div.body"):
                rating_span = review_body.select_one("div.content-reactions-strip span.rating")
                rating = None
                if rating_span:
                    for cls in rating_span.get("class", []):
                        if cls.startswith("rated-"):
                            try:
                                rating = int(cls.split("rated-")[1]) / 2
                                break
                            except:
                                rating = None

                review_text_div = review_body.select_one("div.js-review div.body-text")
                review_text = review_text_div.text.strip() if review_text_div else None

                if review_text:
                    sentiment = get_sentiment(review_text)
                    reviews.append({
                        "movie_slug": slug,
                        "review": review_text,
                        "stars": rating,
                        "sentiment": sentiment
                    })
                    total_reviews_collected += 1

                if total_reviews_collected >= REVIEWS_PER_MOVIE:
                    break

            # Go to next page if needed
            if total_reviews_collected >= REVIEWS_PER_MOVIE:
                break
            next_page = soup.select_one("div.pagination a.next")
            if next_page and next_page.get("href"):
                current_url = base_url + next_page["href"]
            else:
                break

        await browser.close()

    print(f"✅ {slug}: {total_reviews_collected} reviews collected.")
    return movie_data, reviews


async def main():
    movie_slugs = await get_popular_movie_slugs()
    all_movies = []
    all_reviews = []

    for slug in movie_slugs:
        movie_data, reviews = await scrape_movie_data(slug)
        if movie_data:
            all_movies.append(movie_data)
            all_reviews.extend(reviews)

    # Save to CSV
    pd.DataFrame(all_movies).to_csv("movies.csv", index=False)
    pd.DataFrame(all_reviews).to_csv("reviews.csv", index=False)
    print("📦 Data saved to 'movies.csv' and 'reviews.csv'")


await main()
```
